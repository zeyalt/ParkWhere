## Import modules

```python
from datetime import datetime, timedelta

import datetime as dt
import pandas as pd
import requests
```

## Load dataset

```python
df = pd.read_excel("data/parking_data.xls")
df.head()
```

```python
print(f"Total number of records: {len(df)}")
print("Breakdown of parking zones:\n")
print(df.parking_zone.value_counts())
```

## Feature engineering

```python
# Extract various features from `date_time`
df['year'] = df['date_time'].dt.year
df['month'] = df['date_time'].dt.month
df['day'] = df['date_time'].dt.day
df['day_of_week'] = df['date_time'].dt.weekday
df['hour'] = df['date_time'].dt.hour
df['minute'] = df['date_time'].dt.minute
df['date'] = df['date_time'].dt.date
df['time'] = df['date_time'].dt.strftime('%H:%M')
df['hour_min'] = round(df['hour'] + (df['minute'] / 60), 1)

# Replace numbers by names of the months
df['month'].replace({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
                     7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}, inplace=True)

# Replace numbers by days of the week
# df['day_of_week'].replace({0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}, inplace=True)
    
# Rearrange columns
df = df[['date', 'time', 'year', 'month', 'day', 'day_of_week', 'hour', 'minute', 'hour_min', 'parking_zone']] 

# Sanity check
df.head()
```

Availability of parking lots could also be influenced by whether a particular date is the eve of a public holiday, or a public holiday itself. Hence, we will create another feature to indicate 1 if it is an eve or a public holiday, and a 0 otherwise.

```python
url = 'https://data.gov.sg/api/action/datastore_search?resource_id=550f6e9e-034e-45a7-a003-cf7f7e252c9a&' #2021 PH
# url = 'https://data.gov.sg/api/action/datastore_search?resource_id=04a78f5b-2d12-4695-a6cd-d2b072bc93fe&' #2022 PH
response = requests.get(url)
data = response.json()
pub_hols = [ele['date'] for ele in data['result']['records']]
print(pub_hols)
```

```python
def parse_date(date_string):
    return datetime.strptime(date_string, '%Y-%m-%d').date()

parse_date(pub_hols[1])
```

```python
pub_hols_parsed = [parse_date(ph) for ph in pub_hols]
pub_hols_parsed
```

```python
eve_parsed = [parse_date(ph) - timedelta(days=1) for ph in pub_hols]
eve_parsed
```

```python
import numpy as np

df['ph'] = np.where(df['date'].isin(pub_hols_parsed), 1, 0)
df['eve'] = np.where(df['date'].isin(eve_parsed), 1, 0)
df
```

Convert `month` column into ordered categorical data type

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['month'] = df['month'].astype('category') 
df['month'].cat.set_categories(new_categories=month_order, ordered=True, inplace=True)
df
```

Convert `day_of_week` into ordered categorical data type

```python
day_order = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'] 
df['day_of_week'] = df['day_of_week'].astype('category') 
df['day_of_week'].cat.set_categories(new_categories=day_order, ordered=True, inplace=True)
df
```

Convert `hour` into ordered categorical data type

```python
hour_order = list(range(24))
df['hour'] = df['hour'].astype('category') 
df['hour'].cat.set_categories(new_categories=hour_order, ordered=True, inplace=True)
df
```

```python
df.info()
```

```python
features = ['day_of_week', 'hour_min', 'ph', 'eve']
features
```

```python
train_df = df[features + ['parking_zone']]
train_df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Step 2: Make an instance of the Model
clf = DecisionTreeClassifier(max_depth = 3, random_state = 0, criterion = "entropy")
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_df = pd.get_dummies(train_df, columns=['day_of_week'], prefix='')
```

```python
X = train_df.drop(['parking_zone'], axis=1)
y = train_df[['parking_zone']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clf.fit(X_train, y_train)
```
